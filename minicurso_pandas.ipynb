## 📦 **O que é Pandas?**

Pandas é uma biblioteca open-source do Python voltada para **data manipulation and analysis**.
É amplamente utilizada em ciência de dados e vigilância genômica para trabalhar com dados estruturados.

Principais estruturas:

- `Series`
- `DataFrame`



## 📊 **O que é um DataFrame?**

Um `DataFrame` é uma estrutura tabular bidimensional com linhas e colunas rotuladas.

Em vigilância genômica, cada linha pode representar uma sequência genômica,
e cada coluna representa um metadado epidemiológico ou técnico.


## 📌 **Exemplo de DataFrame**

```python

import pandas as pd

data = {
    'id_sequencia': ['SEQ001', 'SEQ002', 'SEQ003'],
    'patogeno': ['SARS-CoV-2', 'SARS-CoV-2', 'DENV-1'],
    'linhagem': ['BA.1', 'BA.2', 'Genotipo I'],
    'estado': ['MG', 'SP', 'RJ'],
    'ano_coleta': [2023, 2024, 2024],
    'cobertura_pct': [98.5, 96.2, 92.1]
}

df = pd.DataFrame(data)
df
```

## 🔍 **Explorando um DataFrame**

```python

df.head()
df.info()
df.describe()
```

## **🎯 Acessando & Filtrando Dados**

```python

df['linhagem']
df[['patogeno', 'estado']]
df[df['ano_coleta'] > 2023]
df[(df['patogeno'] == 'SARS-CoV-2') & (df['estado'] == 'SP')]
```

## **🔁 Operações CRUD no Pandas**

```python

# Create
new_row = pd.DataFrame([{
    'id_sequencia': 'SEQ004',
    'patogeno': 'SARS-CoV-2',
    'linhagem': 'XBB.1',
    'estado': 'MG',
    'ano_coleta': 2025,
    'cobertura_pct': 97.4
}])

df = pd.concat([df, new_row], ignore_index=True)

# Read
df.loc[df['linhagem'] == 'BA.2']

# Update
df.loc[df['id_sequencia'] == 'SEQ002', 'cobertura_pct'] = 97.8

# Delete
df = df[df['cobertura_pct'] >= 95]
```

## 🧹 **Limpeza de Dados**

```python

df.isnull().sum()
df.fillna({'cobertura_pct': 0}, inplace=True)
```

## 📊 **Análise de Dados**

```python

df['linhagem'].value_counts()
df.groupby('estado')['id_sequencia'].count()
df.sort_values(by='ano_coleta', ascending=False)
```

## 💾 **Salvando Dados**

```python

df.to_csv('vigilancia_genomica_tratada.csv', index=False)
```
